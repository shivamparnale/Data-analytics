### Bar plot Matplotlib

```python
import matplotlib.pyplot as plt

y = [34, 25, 36, 40, 47]
x = ["Part1", "Part2", "Part3", "Part4", "Part5"]
color = ["red", "green", "yellow", "blue", "orange"]

plt.bar(x, y, color=color, edgecolor="black", width=0.8, bottom=0.8)
plt.xlabel("Parts of Harry Potter", fontsize=12)
plt.ylabel("Popularity", fontsize=12)
plt.title("Popularity of Different Parts of Harry Potter", fontsize=12)
plt.savefig("bar.png")
plt.show()
```

### Stack plot

### Legend matplotlib

```python
x = [1,2,3,4,5]
y= [45,67,33,67,12]
y1 = [41,60,13,66,13]
y2 = [11,30,55,60,19]

plt.figure(figsize=[5,3])
plt.plot(x,y,label="Male")
plt.plot(x,y1,label="Female")
plt.plot(x,y2,label="Kids")
plt.legend(["a1","a2","a3"],ncols=2)
# plt.legend(loc = 1)
# plt.legend(loc = 2)
# plt.legend(loc = 3)
# plt.legend(loc = 4)
# plt.legend(loc = 10)
plt.show()
```

```python
x = [1,2,3,4,5]
y= [45,67,33,67,12]
y1 = [41,60,13,66,13]
y2 = [11,30,55,60,19]
y3 = [11,40,45,90,59]

plt.figure(figsize=[5,3])
plt.plot(x,y,label="Male")
plt.plot(x,y1,label="Female")
plt.plot(x,y2,label="Kids")
plt.plot(x,y3,label="blanks")
plt.legend(bbox_to_anchor = (0.8,1.2),ncols=2,labelspacing=2)
plt.show()
```
